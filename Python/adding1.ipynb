```python
State = []
Party = []
Year = []
Manifesto = []
Source = []
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
State.append('Tripura')
Party.append('BJP')
Year.append('2018')
Source.append('https://www.oneindia.com/india/tripura-election-2018-bjps-vision-document-2642200.html')
html_text=requests.get('https://www.oneindia.com/india/tripura-election-2018-bjps-vision-document-2642200.html').text
print(html_text)
```

```python
State[-1]='UP'
Party[-1]='BJP'
Year[-1]='2017'
Source[-1]='https://www.ndtv.com/india-news/live-bjp-releases-manifesto-for-uttar-pradesh-elections-2017-1653584'
```

```python

html_text=requests.get('https://www.ndtv.com/india-news/live-bjp-releases-manifesto-for-uttar-pradesh-elections-2017-1653584').text
print(html_text)
```

```python
soup = BeautifulSoup(html_text, 'lxml')
manifesto = ''
i =  soup.find("div",{"class" : "leftpanel"})
j = i.find_all("p")
for k in j:
    manifesto += k.text.replace('\n',' ')

    print(k.text)
```

```python
Manifesto.append(manifesto)
```
